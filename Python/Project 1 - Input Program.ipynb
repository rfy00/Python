```python
# Desc: This program uses OOP to add the price to the balance and count the item
#       prices that were entered
# Usage: The program asks a user to enter a price of an item or type "No".
#        Once a price for an item is entered, it adds the price to the balance and
#        counts the items using a class and defining functions

import locale
locale.setlocale(locale.LC_ALL, 'en_US')

# Prints heading to the new file
print('Welcome!\n')
print('This program requests a user to enter a price of an item or type No to end program. ')
print('Once a user types in NO, then the program prints out the dollar balance and item count\n')

class CashRegister:
    def __init__(self):
        self.balance = 0
        self.itemcnt = 0

    def add_item(self, price):

        self.balance += price
        self.itemcnt += 1

    def get_total(self):

        return self.balance

    def get_count(self):

        return self.itemcnt

    def clear(self):
        self.balance = 0

cr = CashRegister()

user_input = ''

while user_input.upper() != 'NO':
    try:  # Error handling for entering item price until user types in "NO"
        user_input = input('Enter an item price (example: 12.56) or type NO to end the program: ')
        cr.add_item(price=float(user_input))

        if not user_input.upper():
            raise ValueError

    except ValueError:
        if user_input.upper() == 'NO':    # "No" ends the program and prints the information
            print('--------------------------------------------------------------------------------')
            print('\nThe total dollar amount of items is: ' + str(locale.currency(cr.get_total(), grouping=True)))
            print('The total number of items entered is: ' + str(cr.get_count()))
            cr.clear()
        else:    # This handles the issue with a user not selecting "No" or a float amount
            print('\nThe number was entered incorrectly')
            user_input = input('Enter an item price (example: 12.56) or type NO to end the program: ')
            cr.add_item(price=float(user_input))
```
